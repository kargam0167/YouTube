```python
import json
import os
import codecs
from tqdm.notebook import tqdm
source_dataset_dir = 'CHRISFREELANSER/CHRIS_FREELANSER_HOWTO_BE.json'
list_json = []
with codecs.open(source_dataset_dir, "r", "utf-8") as corpus_file:
	for idx, line in enumerate(tqdm(corpus_file)):
		json_ = json.loads(line)
		list_json.append(json_)
print (list_json[0])
print (list_json[0]['title'])
```
